Write a function that displays a HTML that displays the following information for a given list of numbers:
---------------------------------------------------------------------------------------------------------------------------

Count: the amount of items in the list.

Minimum: the smallest item in the list.

Maximum: the largest item in the list.

Median: the item halfway in the list after ordering from least to greatest.

Mode: the item that appears the most often in the list.

Mean: the average of the items in the list.

Total: the total sum of the items int the list.

Data: the list itself.

The list of numbers will be given to the program via Jupyter's interaction module in the form of a string. To separate individual items in the list, the user may select between any of the following separators: ' ', ',', ':', '|'.


```python
### from IPython.display import HTML, display
from IPython.html.widgets.interaction import interact
#from tabulate import tabulate
from IPython.display import HTML, display
from statistics import median
# Functions

def calculate_mode(data):
    ''' Return the mode of a list of numbers '''
    d = dict()
    for i in data:
        if i not in d:
            d[i] = 1
        else:
            d[i] = d[i] + 1        
#    print d
    mode = max(d, key=d.get)       
    return mode

def print_statistics(raw_data, separator):
    ''' Calculate and display the statistics of the input data '''
    if len(raw_data) == 0:
        raw_data = raw_data.replace('','0')
    data = raw_data.split(separator)   
    
    data = [int(s) for s in data]
    Count = len(data)
    data.sort()
    Minimum = min(data)
    Maximum = max(data)
    Median = median(data)
    
    Total = 0
    for i in data:
        Total += i    

    Mean = float(Total)/Count
    Mode = calculate_mode(data)
#    table = [[Count,Minimum,Maximum,Median,Mode,Mean,Total,data]]  
#    print tabulate(table, headers=["Count","Minimum","Maximum","Median","Mode","Mean","Total","Data"],
#                   tablefmt="fancy_grid")
    html = '''
            <table>
              <thead>
                <th>Count</th>
                <th>Minimum</th>
                <th>Maximum</th>
                <th>Median</th>
                <th>Mode</th>
                <th>Mean</th>
                <th>Total</th>
                <th>Data</th>
              </thead>
              <tbody>
                <tr>
                <td>{}</td>
                <td>{}</td>
                <td>{}</td>
                <td>{}</td>
                <td>{}</td>
                <td>{}</td>
                <td>{}</td>
                <td>{}</td>
                </tr>
              </tbody>
            </table>
        '''.format(Count,Minimum,Maximum,Median,Mode,Mean,Total,data)
    display(HTML(html))
# Run interactive loop

interact(print_statistics, raw_data='', separator=(' ', ',', ':', '|'))
```
