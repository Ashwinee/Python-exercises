Google has an excellent web service that allows us to make use of their large database of geographic information. Geocoding is the process of converting addresses (like "1600 Amphitheatre Parkway, Mountain View, CA") into geographic coordinates (like latitude 37.423021 and longitude -122.083739), which you can use to place markers on a map, or position the map.The following is a simple application to prompt the user for a search string, call the Google geocoding API, and extract information from the returned JSON.

For detail documentation of Google Maps Geocoding API, visit  
https://developers.google.com/maps/documentation/geocoding/intro#reverse-place-id

For details of urllib, visit https://docs.python.org/2/library/urllib.html

```python
import urllib
import json

serviceurl = 'http://maps.googleapis.com/maps/api/geocode/json?'

while True:
    address = raw_input('Enter location: ')
    if len(address) < 1 : break

    url = serviceurl + urllib.urlencode({'sensor':'false', 
          'address': address})
    print 'Retrieving', url
    geocoding = urllib.urlopen(url)
    data = geocoding.read()
    print 'Retrieved',len(data),'characters'

    try: js = json.loads(str(data))
    except: js = None
    if 'status' not in js or js['status'] != 'OK':
        print '==== Failure To Retrieve ===='
        print data
        continue

    print json.dumps(js, indent=4)
      
    lat = js["results"][0]["geometry"]["location"]["lat"]
    lng = js["results"][0]["geometry"]["location"]["lng"]
    print 'lat',lat,'lng',lng
    location = js['results'][0]['formatted_address']
    print location
```
